```python
import pandas
import re
df = pandas.read_excel('chongqing2.xlsx')
df.head()
```

#### 单个因素与价格的关系

```python
#按楼层统计其各自的销售总价均值并进行降序排序。目的：房屋楼层与房屋价格之间的关系
floor = df.groupby('楼层')['总价'].mean().sort_values(ascending=False)
print(floor)
```

```python
#图形
floor.plot(kind='bar',x='楼层',y='总价',figsize=[10,5],fc='b')  
plt.grid()#设置网格线
plt.title('楼层与总价的关系',fontsize=14)#设置标题
#设置坐标轴标签
plt.xlabel('楼层',fontsize=14)
plt.ylabel('总价均值',fontsize=14)
#设置刻度标签参数大小
plt.tick_params(labelsize=15)
plt.show()
```

```python
#按装修统计其各自的总价均值并将求得的数据进行降序排序
decoration = df.groupby('装修')['总价'].mean().sort_values(ascending=False)
print(decoration)
```

#### 多个因素与总价的关系

```python
df2 = df.pivot_table(index=['产权性质'],columns='装修',values='总价',aggfunc='mean',fill_value=0) #表中数据为对应交叉对象的总价之和
df2
```

```python
df3 = df.pivot_table(index=['朝向'],columns='楼层',values='总价',aggfunc='mean',fill_value=0) #表中数据为对应交叉对象的总价之和
df3
```

```python
df4 = df.pivot_table(index=['户型'],columns='建造年数',values='总价',aggfunc='mean',fill_value=0) #表中数据为对应交叉对象的总价之和
df4
```

```python
df5 = df.pivot_table(index=['面积','装修'],columns='楼层',values='总价',aggfunc='mean',fill_value=0) #表中数据为对应交叉对象的总价之和
df5
```
