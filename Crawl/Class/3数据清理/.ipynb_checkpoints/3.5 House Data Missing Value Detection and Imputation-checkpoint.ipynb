## 补齐缺失值

```python
df.head(3)
```

```python
#对物业费为空的单元格，填补为0
df['物 业 费'] = df['物 业 费'].fillna(0)
```

```python
df.head(3)
```

```python
#计算'總價'列为空的单元格个数
df['總價'].isnull().sum()
```

```python
#筛选出‘總價’为空的行记录
df[df['總價'].isnull()]
```

```python
#利用‘總價’列与‘建筑面积’列求均价
(df['總價'] / df['建筑面积']).mean()
```

```python
df['總價'].isnull().sum()
```

```python
#对‘總價’列单元格为空的，按上面求的“均价”与建筑面积相乘进行计算
df['總價'] = df['總價'].fillna((df['總價'] / df['建筑面积']).mean()* df['建筑面积'])
```

```python
#再次计算‘總價’为空的单元格个数
df['總價'].isnull().sum()
```
