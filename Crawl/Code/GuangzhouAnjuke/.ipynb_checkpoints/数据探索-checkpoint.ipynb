```python
import pandas
#读取csv文件数据
df = pandas.read_csv('house_final.csv',index_col=0)
df.head()
```

```python
#按总价进行降序排序
df2 = df.sort_values(by="总价" , ascending=False).head(10)
df2.head()
```

```python
#对df2各非字符串列进行描述性统计（计数，均值，标准差，最大最小，4分位等）
df2.describe()
```

```python
#按客户统计他们的消费总金额，并按降序排列,最终取前10个用户的统计数据
g = df2['总价'].sort_values(ascending=False)[0:10]
#绘制前10个用户的消费金额柱形图，图片大小10*5
g.plot(kind='bar',figsize=[10,5],color='blue')
```

```python
import matplotlib.pyplot as plt

# 中文和负号的正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

price = df2['总价']

# 绘图
plt.bar(range(10), price,align = 'center',color='steelblue', alpha = 0.8)
# 添加轴标签
plt.ylabel('总价/万')
# 添加标题
plt.title('安居客广州二手房总价Top10')
area = df2['面积'].values.tolist()[0:10]
# 添加刻度标签
# plt.xticks(area)
# 设置Y轴的刻度范围
plt.ylim([100,2000])

# 为每个条形图添加数值标签
for x,y in enumerate(price):
    plt.text(x,y+100,'%s' %round(y,1),ha='center')
# 显示图形
plt.show()
```

```python
#按总价进行升序排序
df3 = df.sort_values(by="总价" , ascending=True).head(10)
df3.head()
```

```python
#按客户统计他们的消费总金额，并按降序排列,最终取前10个用户的统计数据
g2 = df3['总价'].sort_values(ascending=True)[0:10]
#绘制前10个用户的消费金额柱形图，图片大小10*5
g2.plot(kind='bar',color='blue')

# 中文和负号的正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

price = df3['总价']

# 绘图
plt.bar(range(10), price,align = 'center',color='steelblue', alpha = 0.8)
# 添加轴标签
plt.ylabel('总价/万')
# 添加标题
plt.title('安居客广州二手房总价最低Top10')
area = df3['面积'].values.tolist()[0:10]
# 添加刻度标签
# plt.xticks(area)
# 设置Y轴的刻度范围
# plt.ylim([25,80])

# # 为每个条形图添加数值标签
# for x,y in enumerate(price):
#     plt.text(x,y+100,'%s' %round(y,1),ha='center')
# 显示图形
plt.show()
```

```python
#按楼层统计其各自的销售总价均值并进行降序排序。房屋楼层与房屋价格之间的关系
floor = df.groupby('楼层')['总价'].mean().sort_values(ascending=False)
floor
```

```python
floor.plot(kind='bar',x='楼层',y='总价',figsize=[10,5],fc='b',rot=45)  
plt.grid()#设置网格线
plt.title('楼层与总价的关系',fontsize=14)
#设置坐标轴标签
plt.xlabel('楼层',fontsize=14)
plt.ylabel('总价均值',fontsize=14)
#设置刻度标签参数大小
plt.tick_params(labelsize=15)
plt.show()
```

```python
#按户型统计其各自的总价均值并将求得的数据进行降序排序
Housetype = df.groupby('户型')['总价'].mean().sort_values(ascending=False)
Housetype
```

```python
Housetype.plot(kind='bar',x='户型',y='总价',figsize=[12,7],fc='orange',rot=45)  
# plt.grid()#设置网格线
plt.title('户型与总价的关系',fontsize=14)
#设置坐标轴标签
plt.xlabel('户型',fontsize=14)
plt.ylabel('总价均值',fontsize=14)
#设置刻度标签参数大小
plt.tick_params(labelsize=15)
plt.show()
```

```python
#按户型统计其各自的总价均值并将求得的数据进行降序排序
layout = df.groupby('装修')['总价'].mean().sort_values(ascending=False)
layout.plot(kind='bar',x='装修',y='总价',figsize=[12,7],fc='r',rot=45)  
# plt.grid()#设置网格线
plt.title('装修与总价的关系',fontsize=14)#设置标题
#设置坐标轴标签
plt.xlabel('装修',fontsize=14)
plt.ylabel('总价均值',fontsize=14)
#设置刻度标签参数大小
plt.tick_params(labelsize=15)
plt.show()
```

```python
#按装修统计其各自的总价均值并将求得的数据进行降序排序
dec= df.groupby('装修')['总价'].mean().sort_values(ascending=False)
dec
```

```python
dec.plot(kind='bar',x='装修',y='总价',figsize=[10,5],fc='r',rot=45)  
plt.grid()#设置网格线
plt.title('装修与总价的关系',fontsize=14)
#设置坐标轴标签
plt.xlabel('装修',fontsize=14)
plt.ylabel('总价均值',fontsize=14)
#设置刻度标签参数大小
plt.tick_params(labelsize=15)
plt.show()
```
